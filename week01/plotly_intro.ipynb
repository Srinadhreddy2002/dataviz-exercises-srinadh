```python
import plotly.express as px
import plotly.io as pio

# This ensures the graph opens properly
pio.renderers.default = "browser"

# Load sample stock data
df = px.data.stocks(datetimes=True)

# Create the line chart with red color
fig = px.line(
    df,
    x="date",
    y="GOOG",
    title="Google Stock Price",
    color_discrete_sequence=["red"]
)

# Show the graph
fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

# Open graph in browser
pio.renderers.default = "browser"

# Load dataset
df = px.data.stocks(datetimes=True)

# Convert data to long format (needed for multiple lines)
df_long = df.melt(id_vars="date", var_name="Company", value_name="Price")

# Create multi-line chart
fig = px.line(
    df_long,
    x="date",
    y="Price",
    color="Company",
    title="Stock Prices of Multiple Companies"
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

# keep this so it opens properly
pio.renderers.default = "browser"

# load data
df = px.data.stocks(datetimes=True)

# convert for multiple companies
df_long = df.melt(id_vars="date", var_name="Company", value_name="Price")

# new chart with markers (dots)
fig = px.line(
    df_long,
    x="date",
    y="Price",
    color="Company",
    title="Stock Prices with Markers",
    markers=True
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

df = px.data.stocks(datetimes=True)

# Take last available data
last = df.tail(1).melt(id_vars="date", var_name="Company", value_name="Price")

fig = px.bar(
    last,
    x="Company",
    y="Price",
    title="Latest Stock Prices"
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

df = px.data.stocks(datetimes=True)

# Take latest values
last = df.tail(1).melt(id_vars="date", var_name="Company", value_name="Price")

fig = px.pie(
    last,
    names="Company",
    values="Price",
    title="Stock Price Distribution"
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

# Real dataset: gapminder (countries data)
df = px.data.gapminder()

# Filter for one year
df_2007 = df[df["year"] == 2007]

fig = px.scatter(
    df_2007,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    title="GDP vs Life Expectancy (2007)"
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

df = px.data.gapminder()

fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    size_max=60,
    animation_frame="year",
    animation_group="country",
    title="GDP vs Life Expectancy Over Time"
)

fig.show()
```

```python
import plotly.express as px
import plotly.io as pio

pio.renderers.default = "browser"

# Load real dataset
df = px.data.gapminder()

# Take latest year
latest = df[df["year"] == 2007]

# Get top 10 countries by life expectancy
top10 = latest.sort_values(by="lifeExp", ascending=False).head(10)

fig = px.bar(
    top10,
    x="country",
    y="lifeExp",
    color="continent",
    title="Top 10 Countries by Life Expectancy (2007)"
)

fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder()

# Take one year
df_2007 = df[df["year"] == 2007]

fig = px.histogram(
    df_2007,
    x="lifeExp",
    nbins=20,
    title="Distribution of Life Expectancy (2007)"
)

fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder()

# Filter one year
df_2007 = df[df["year"] == 2007]

fig = px.box(
    df_2007,
    x="continent",
    y="lifeExp",
    title="Life Expectancy Distribution by Continent (2007)"
)

fig.show()
```
