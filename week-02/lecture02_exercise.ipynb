# World Happiness Report 2023 Analysis

This notebook analyzes global happiness data using Plotly.

```python
import pandas as pd

df = pd.read_csv("world_happiness_2023.csv", sep=';')

df.columns = df.columns.str.strip().str.replace(" ", "_")

for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
        df[col] = pd.to_numeric(df[col], errors='ignore')

df.head()
```

```python
# Force convert important columns to numeric
df['Happiness_score'] = df['Happiness_score'].astype(str).str.replace(',', '.', regex=False)
df['Happiness_score'] = pd.to_numeric(df['Happiness_score'], errors='coerce')
```

## Task 1 — Regional Comparison

This chart shows the average happiness score across different regions.

```python
import plotly.express as px

# Average happiness by region
region_avg = df.groupby('Regional_indicator')['Happiness_score'].mean().reset_index()

# Sort for better visualization
region_avg = region_avg.sort_values('Happiness_score')

# Bar chart
fig1 = px.bar(
    region_avg,
    x='Happiness_score',
    y='Regional_indicator',
    orientation='h',
    text=region_avg['Happiness_score'].round(2),
    color='Happiness_score',
    color_continuous_scale='Blues'
)

fig1.update_layout(
    title="Average Happiness Score by Region",
    xaxis_title="Happiness Score",
    yaxis_title="Region",
    template='simple_white'
)

fig1.show()
```

Western Europe and North America show the highest happiness levels, while South Asia and Sub-Saharan Africa have lower scores.

## Task 2 — Top vs Bottom Countries

This chart compares the happiest and least happy countries.

```python
# Top 8 happiest countries
top8 = df.nlargest(8, 'Happiness_score')
top8['Group'] = 'Top 8'

# Bottom 8 countries
bottom8 = df.nsmallest(8, 'Happiness_score')
bottom8['Group'] = 'Bottom 8'

# Combine
combined = pd.concat([top8, bottom8]).sort_values('Happiness_score')

# Global average
global_avg = df['Happiness_score'].mean()

# Plot
fig2 = px.bar(
    combined,
    x='Happiness_score',
    y='Country',
    orientation='h',
    color='Group',
    text=combined['Happiness_score'].round(2),
    color_discrete_map={
        'Top 8': 'green',
        'Bottom 8': 'red'
    }
)

fig2.update_layout(
    title="Top vs Bottom Happiness Countries",
    xaxis_title="Happiness Score",
    yaxis_title="Country",
    template='simple_white'
)

# Add average line
fig2.add_vline(
    x=global_avg,
    line_dash="dash",
    line_color="black",
    annotation_text=f"Avg: {global_avg:.2f}",
    annotation_position="top"
)

fig2.show()
```

There is a clear gap between the happiest and least happy countries, highlighting global inequality in well-being.
