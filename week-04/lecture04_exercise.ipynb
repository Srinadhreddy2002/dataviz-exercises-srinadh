Task 1 — Scatter: life expectancy change over time

```python
# Task 1

import pandas as pd
import plotly.express as px

df = px.data.gapminder()

# filter: one continent (Asia) + two years
df_task1 = df.loc[(df['continent'] == 'Asia') & (df['year'].isin([2000, 2007]))]

fig = px.scatter(
    data_frame=df_task1,
    x='gdpPercap',
    y='lifeExp',
    color='year',                         # only 2 colours
    hover_name='country',
    log_x=True,                           # log scale required
    labels={
        'gdpPercap': 'GDP per Capita (log scale)',
        'lifeExp': 'Life Expectancy',
        'year': ''
    },
    color_discrete_sequence=['#4CC9F0', '#F72585']   # simple 2 colours
)

fig.update_layout(
    title='Asia: Countries moved toward higher wealth and longer lives (2000 → 2007)',
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    xaxis=dict(gridcolor='#EEEEEE'),
    yaxis=dict(gridcolor='#EEEEEE'),
    margin=dict(l=60, r=40, t=55, b=40),
    height=500
)

fig.update_traces(marker=dict(size=9, opacity=0.7))

fig.show()
```

Task 2 — Bubble Chart (Highlight ONE continent)

```python
# Task 2

import math

df_2007 = df[df['year'] == 2007]

highlight_continent = 'Asia'

# grey + highlight map
colour_map = {
    c: ('#E63946' if c == highlight_continent else '#DDDDDD')
    for c in df_2007['continent'].unique()
}

fig = px.scatter(
    data_frame=df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop',
    color='continent',
    color_discrete_map=colour_map,
    hover_name='country',
    log_x=True,
    size_max=60,                          # REQUIRED
    labels={
        'gdpPercap': 'GDP per Capita (log scale)',
        'lifeExp': 'Life Expectancy',
        'pop': 'Population',
        'continent': ''
    },
    title='Asia: Large populations with improving wealth and life expectancy'
)

fig.update_traces(
    showlegend=False,
    opacity=0.7
)

# highlight styling
fig.update_traces(
    selector=dict(name=highlight_continent),
    marker=dict(size=10, opacity=0.85)
)

fig.update_traces(
    selector=lambda t: t.name != highlight_continent,
    marker=dict(size=7, opacity=0.5)
)

# annotation (important!)
fig.add_annotation(
    x=math.log10(5000),
    y=70,
    text='Asia: rising economies<br>with large populations',
    showarrow=False,
    font=dict(color='#E63946', size=11, family='Arial'),
    bgcolor='white',
    bordercolor='#E63946',
    borderwidth=1,
    borderpad=4
)

fig.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font=dict(family='Arial', size=12),
    xaxis=dict(gridcolor='#EEEEEE'),
    yaxis=dict(gridcolor='#EEEEEE'),
    margin=dict(l=60, r=40, t=55, b=40),
    height=600,
    width=1100
)

fig.show()
```
